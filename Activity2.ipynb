```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import re

urls = [
    'https://facts.net/characters/20-fun-facts-about-captain-america/',
    'https://www.beano.com/posts/captain-america-facts',
    'https://www.watchmojo.com/articles/top-10-captain-america-facts',


    ]

def clean_text(text):
  unwanted_pattern = [
      r'Key Takeaways:',
      r'Conclusion',
      r'FAQs',
      r'Test your Captain America trivia now!',
      r'More Stuff',
      r'RELATED'
      ]
  for pattern in unwanted_pattern:
    text = re.sub(pattern, '', text)
  return text.strip()

def extract_data_from_url(url):
  response = requests.get(url)

  if response.status_code == 200:
    soup = bs(response.text, 'html.parser')
    facts = []

    h2_tags = soup.find_all('h2')

    for h2 in h2_tags:
      h2_text = h2.get_text(strip=True)
      cleaned_text = clean_text(h2_text)

      if cleaned_text == '':
        continue
      facts.extend(cleaned_text.split('\n'))


  return {
      'URL':url,
      'Facts':facts
  }
all_facts = []
URL = []
for url in urls:
  data = extract_data_from_url(url)
  for facts in data['Facts']:
    URL.append(data['URL'])
    all_facts.append(facts)

data = {
    "URL":URL,
    'Facts':all_facts
}


df_facts = pd.DataFrame(data)
df_facts.to_csv('captain_america_facts.csv', index=False)
print(df_facts)

```

# TRYING ANOTHER WEBSITE TO SCRAPE

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
import random
import time

urls = [
    'https://marvelcinematicuniverse.fandom.com/wiki/Steve_Rogers#Trivia'
]

def clean_text(text):
    unwanted_pattern = [
        r"#\d+"
    ]
    for pattern in unwanted_pattern:
        text = re.sub(pattern, '', text)
    return text.strip()

def extract_data_from_url(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

    # Send a GET request to the page
    try:
        response = requests.get(url, headers=headers)

        if response.status_code == 200:
            soup = bs(response.text, 'html.parser')
            facts = []

            trivia_header = soup.find('span', id='Trivia')

            if trivia_header:
                ul_tag = trivia_header.find_next('ul')

                if ul_tag:
                    li_tags = ul_tag.find_all('li')

                    for li in li_tags:
                        li_text = li.get_text(strip=True)
                        cleaned_text = clean_text(li_text)

                        if cleaned_text == '':
                            continue
                        facts.append(cleaned_text)

            return {
                'URL': url,
                'Facts': facts
            }
    except requests.ConnectionError as e:
        print(f"Connection error occurred: {e}")
        return None
    finally:
        # Wait for a random time between 1 and 5 seconds
        time.sleep(random.uniform(1, 5))

all_facts = []
URL = []

for url in urls:
    data = extract_data_from_url(url)
    if data and 'Facts' in data:
        for fact in data['Facts']:
            URL.append(data['URL'])
            all_facts.append(fact)

data = {
    "URL": URL,
    'Facts': all_facts
}

df_facts_02 = pd.DataFrame(data)
df_facts_02.to_csv('captain_america_MARVELCINEMATIC_facts.csv', index=False)
print(df_facts_02)
```
